## V. Evaluation of Significance <a name="evaluationOfSignificance"></a>

### Main Evaluation I — Diabetes <a name="main1"></a>

### 1. Linear Regression:  "no_exercise", "smoker" and "few_fruit_veg" vs. "diabetes"

```python
#create Linear Regression models for variables we are interested in investigating:

#Declare independent and dependent variables
independent_vars = ["no_exercise","smoker", "few_fruit_veg"]
dependent_var = "diabetes"

for s in independent_vars:
    model = LinearRegression().fit(df[[s]],df[dependent_var])
    print(f"column: {s}")
    print(f"slope: {round(model.coef_[0],3)}")
    print(f"intercept: {round(model.intercept_,3)}")
    r2 = round(model.score(df[[s]],df[dependent_var]),5)
    print(f"R-squared: {r2}")
    print(f"The correlation between '{dependent_var}' and '{s}': {round(df[dependent_var].corr(df[s]),3)}\n")
    
    #plot each observations
    plt.scatter(df[s], df[dependent_var], alpha=0.2)
    #plot line of best fit
    plt.plot(df[s], df[s]*model.coef_ + model.intercept_,label = s)
    
#plot properties   
plt.title("diabetes vs no_exercise, smoker and few_fruit_veg")
plt.xlabel("percentage of county residents")
plt.ylabel("Percentage of residents with diabetes")
plt.legend()
plt.show()
```

### Analysis:

The R-squared values for no exercise, smoker, and few vegatables and fruits range from around 0.035 to 0.33. This means that the about 33% of the variation of diabetes can be explained by difference in the proportion of people who exercises. On the other hand, only 3.5% of the variation in diabetes can be explained by consumption of fruits and vegetables, meaning that the diabetes has many other impacting factors that is not incorporated into our model. For reference, the percentage variation in diabetes (33%) that can be explained by variation in no exericing is more than nine times of that (3.5%) explained by variation in low fruits and vegetable intake.

All three independent variables have a positive slope. The values indicate that for every percentage increase in the people who don't exercise in a given county, the percent of people who have diabetes is expected to grow by 0.196%. The same logic applies to the other two. For every percentage increase in the people who smokes in a given county, the percent of people who have diabetes is expected to grow by 0.193%. For every percentage increase in the people who don't exercise in a given county, the percent of people who have diabetes is expected to grow by 0.085%. The slopes for smoker and no exercise are quite similar, and both are larger than low fruits and vegetables. It can be seen that the slopes are relatively small (<0.2), meaning that the model projects that only a big change in the proportion of people conducting these unhealthy behaviors will correspond to significant changes in proportion of population with diabetes.

Notice that the value of R-squred is the square of the coefficient between each of the independent and dependent variables for single linear regression.

### 2. Hypothesis Testing

**Null Hypothesis:** There is no relationship between diabetes and each of no exercise, smoker, and few fruits & vegetable intake. 

**Alternative Hypothesis:** There exists a relationship between diabetes and each of no exercise, smoker, and few fruits & vegetable intake. 

#### 2.1 Define permutation method to visualize the p-value

```python
def permute(array):
    return np.random.permutation(array)
```

#### 2.2 Using permutation method to visualize the p-value

```python
#plot true regression
colors= sns.color_palette()[:3]

for i,s in enumerate(independent_vars):
    model = LinearRegression().fit(df[[s]],df[dependent_var])
    plt.plot(df[s], df[s]*model.coef_ + model.intercept_,label = s,color = colors[i])

#plot random permutation
for i,s in enumerate(independent_vars):
    for _ in range(500):
        shuffled_y = permute(df[dependent_var])
        random_model = LinearRegression().fit(df[[s]],shuffled_y)
        plt.plot(df[[s]],df[s]*random_model.coef_ + random_model.intercept_,color = colors[i], alpha = 0.02)

plt.xlabel("percentage of residents in a county")
plt.ylabel("Percentage of residents within a county with diabates")
plt.legend()
plt.show()
```

**Method explanation:**

We performed the random shuffling permutation test on each of the dataset in hope to visualize the probability of having such distribution at random. We have shuffled the observations for 500 times. All simulated slopes cross the mean value for the x and y variables at this value will never change. The randomized slopes all hover over the horizontal line with slope = 0. It can be seen that our slopes for all three variables are significantly larger in magnitude. One can assume that the p-value will be very close to, if not, zero.

```python
for s in independent_vars:
    print(f"P-value for the regression between {dependent_var} and {s} is: ")
    print(round(stats.linregress(df[s],df[dependent_var]).pvalue,5))
    print("\n")
```

#### 2.3 Insight

Assume that we use a significance level of 95%, which is alpha = 0.05, for the entirety of our data manipulation. The value chosen is rather arbituary. 

As predicted above, all p-values, when rounded to five significant figures, have a value of zero. In reality, it is an incredibly small number. Therefore, we can conclude that the relaitonship between diabetes and each of no exercise, smoker, and few fruits & vegetable intake is significant, meaning it is not likely that the relationship is by chance. We hereby reject the null hypothsis and determine that the true slope value is not zero and there does exist a relationship. 

Notice that for few fruits and vegetables, the R-squared value is incredibly small, indicating a noisy data with high-variability. However, based on the calculated p values, the trendline is in fact significant! This tells us that low R-squared values and low p values are not incompatible. Hence, even though each of the data points are quite far away from the line of best fit, the relationship is still valid. To elaborate, this also signals that given a value for few fruits and vegetables intake, the model will likely not be precise in predicting the percent of people with diabetes. To increase the R-squared value, we need to take into account of more variables. We'll try the multiple regression model below.

### 3. Multiple Regression

```python
print(f"Multiple regression statistics: {dependent_var} vs {independent_vars}:")
multi_model = LinearRegression().fit(df[independent_vars],df[dependent_var])
print(f"slope: {multi_model.coef_}")
print(f"intercept: {round(multi_model.intercept_,3)}")
r2 = round(multi_model.score(df[independent_vars],df[dependent_var]),5)
print(f"R-squared: {r2}")

#p-value
print(f"Statistics for multiple regression between {dependent_var} and {s}")
stats_input = "{} ~ {}".format(dependent_var," + ".join(independent_vars))
ols_model_results = smf.ols(stats_input,data=df).fit()
ols_model_results.summary()
```

**Insight:**
With a R-squared value of 0.369, the multiple regression model best explains the variation in proportion of diabetes population with differences in the three input variables. Having all three independent variables analyzed, we are able to put together some of the confounding factors that might be present in single regressions. In other words, having more variables to test makes the model more precise. However, we might run a risk of overfitting.

One thing to keep in mind is that the model suggested a high multicollinearity, which makes sense as we saw above that the "independent variables" are correlated with one another, making them not so independent. The effect might be the decrease in slope values for all three of the independent variables compare to single regression slopes. This might be a victim of the high multicollinearity. Note here, the slope of few_fruit_veg even changed from positive to negative, signalling that in this model, the more people who eat less fruit and vegetables in a county, the lower the proportion of people who have diabetes.

Again, the extremely low p values (< 0.002) indicate that the relationship is significant.

### 3. Multiple Regression

```python
print(f"Multiple regression statistics: {dependent_var} vs {independent_vars}:")
multi_model = LinearRegression().fit(df[independent_vars],df[dependent_var])
print(f"slope: {multi_model.coef_}")
print(f"intercept: {round(multi_model.intercept_,3)}")
r2 = round(multi_model.score(df[independent_vars],df[dependent_var]),5)
print(f"R-squared: {r2}")

#p-value
print(f"Statistics for multiple regression between {dependent_var} and {s}")
stats_input = "{} ~ {}".format(dependent_var," + ".join(independent_vars))
ols_model_results = smf.ols(stats_input,data=df).fit()
ols_model_results.summary()
```

**Insight:** With an R squared value of 0.418, we determine that in the multiple regression model, 41.8 percent of variation in obesity is exaplined by variation in the three input variables. This R squared value is greater than any of the single regression model, meaning that multiple regression more accurated produce a model that predicts obesity. However, the increase is not substantial and it is evident that we might be overfitting, as in the case for diabetes. This can also be seen in the value of the slope: all three, though remain positive as in the single regression, has decreased by approximately more than half.

The extremely low p values (< 0.000) indicate that the relationship is significant. Therefore, we reject the null.

Our dataset is missing a lot of values in each of the columns we used. The potential effects of missing data on our results are:

    * There were variables with less than 0 values, such as -1111.1 or -2222, which are considered as “no report” or “no data available”; thus, we excluded the rows containing these values.
    * After cleaning out the missing data, we found that although the number of columns decreased by half, the total population included in the dataset only decreased by around 10%. This means cutting off these parts of data did not affect the general trends of our dataset by too much. 
    * Another thing to point out is that losing 10% of the data may cause a decrease in the variance of our dataset. 
* One potential bias is the method of collecting data. It does not say what the stardized method is and so there is room for bias. Regardless, the dataset is provided by CDC, therefore we have sufficient reasons to believe that it should be reliable.
* It should also be taken into account there are potentially other factors that are not included in this data set but is imposing a bias to our results. This data set is extremely big and not at all specific. Our goal is to find correlation and significance among the selected variables to see which variables might be connected to one another more. This is not sufficient to claim causality.
* For example, our analysis of this dataset is overlooking several factors such as income level, poverty rates, demographic(age, gender & race), as well as many others that can have a significant impact on the outcome of each of the variables we are focusing on.
